```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%store -r filtered_df
```

#### Correlation matrix

```python
# Calculate the correlation matrix
correlation_matrix = filtered_df.corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

#### What is the average age of the individuals in the dataset?

```python
average_age = filtered_df['Age'].mean()
print(int(average_age))
```

#### What is the average glucose level for individuals with diabetes and without diabetes?

```python
# Calculate the average glucose level for individuals with diabetes (Outcome=1)
average_glucose_diabetes = filtered_df[filtered_df['Outcome'] == 1]['Glucose'].mean().astype(int)

# Calculate the average glucose level for individuals without diabetes (Outcome=0)
average_glucose_no_diabetes = filtered_df[filtered_df['Outcome'] == 0]['Glucose'].mean().astype(int)

print(f'Average Glucose for individuals with diabetes: {average_glucose_diabetes}')
print(f'Average Glucose for individuals without diabetes: {average_glucose_no_diabetes}')
```

#### What is the average BMI for individuals with diabetes and without diabetes?

```python
# Calculate the average glucose level for individuals with diabetes (Outcome=1)
average_bmi_diabetes = filtered_df[filtered_df['Outcome'] == 1]['BMI'].mean().astype(int)

# Calculate the average glucose level for individuals without diabetes (Outcome=0)
average_bmi_no_diabetes = filtered_df[filtered_df['Outcome'] == 0]['BMI'].mean().astype(int)

print(f'Average BMI for individuals with diabetes: {average_bmi_diabetes}')
print(f'Average BMI for individuals without diabetes: {average_bmi_no_diabetes}')
```

#### How does the distribution of the predictor variables (Glucose and Insulin) differ for individuals with diabetes and without diabetes?

```python
# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# First subplot: Glucose distribution
sns.boxplot(data=filtered_df, x='Outcome', y='Glucose', ax=axes[0])
axes[0].set_title('Distribution of Glucose by Diabetes Status')
axes[0].set_xlabel('Diabetes Status')
axes[0].set_ylabel('Glucose Level')
axes[0].set_xticks([0, 1])
axes[0].set_xticklabels(['No Diabetes', 'Diabetes'])

# Second subplot: Insulin distribution
sns.boxplot(data=filtered_df, x='Outcome', y='Insulin', ax=axes[1])
axes[1].set_title('Distribution of Insulin by Diabetes Status')
axes[1].set_xlabel('Diabetes Status')
axes[1].set_ylabel('Insulin Level')
axes[1].set_xticks([0, 1])
axes[1].set_xticklabels(['No Diabetes', 'Diabetes'])

plt.tight_layout()
plt.show()
```
